```python
def getBits(data_array):
    
    matrix = np.unpackbits(data_array) #dec2bin
    return(np.reshape(np.transpose(matrix),1*8*len(data_array),order = "A")) #reshape transpose matrix


#scratchpad
#import csv 
#with open('testcsv', 'w') as f: 
      
    # using csv.writer method from CSV package 
    #write = csv.writer(f) 
      
   # write.writerow(data) 
    #write.writerows(rows) 
#print(np.unpackbits(data_array))
print(data[:])

# using the savetxt  
# from the numpy module 
#np.savetxt("hello.csv",  
           #data, 
           #delimiter =", ",  
           #fmt ='% s')
```

```python
import math
from scipy.fft import fft, ifft

def phase_encoder(data,data_array):

    L = 1024
    plain =  data[:]
    datafrombits = getBits(data_array)
    I = len(plain)
    m = len(datafrombits)
    N = math.floor(I/L)
    s = np.reshape(plain[:N*L,], L*N, order="A")
    w = fft(s)            # FFT of each segment
    Phi = np.angle(w)        # Phases matrix including each segment
    A = abs(w)         
    

    #Calculating phase differences of adjacent segments
    
    DeltaPhi = np.zeros((L,N))
    
    
    for k in range(2, N):
        DeltaPhi[k]=Phi[k]-Phi[k-1]
        
    #Binary data is represented as {-pi/2, pi/2} and stored in PhiData

    
    PhiData = np.zeros((1,m-1))
    
    for k in range(1, m-1):
        if(datafrombits[k] == '0'):
            PhiData[k] = math.pi/2
        elif(datafrombits[k] == '1'):
            PhiData[k] = -math.pi/2

    
    #PhiData into the middle of first phase matrix
    
   
    Phi[math.floor(L/2-m+1):math.floor(L/2),] = PhiData;        
    Phi[math.floor(L/2+1+1):math.floor(L/2+1+m),] = -np.flip(PhiData); 
    
   #Recreate phase matrix
    
    for k in range(2, N):
        DeltaPhi[k]=Phi[k]-Phi[k-1]
        
    
    #Reconstruct signal in time domain using ifft
    
    z = np.real(ifft((A * np.exp(1j*Phi)))) #Euler's formula
    snew = np.reshape(z, N*L*1, order="A")
    snew = snew + plain[1:N*L+1:]
    return(snew)
```

```python
from scipy.io.wavfile import write
# to be added at end
#out = phase_encoder(data, data_array)

output = phase_encoder(data,data_array)

write("embeddedstego.wav", samplerate, output)
```
