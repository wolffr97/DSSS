```python
def bit_error(hidden,extracted):
    y = hidden
    x = extracted
    
    l = min(len(x),len(y))
    
    ber = 0
    for i in range(0,l):
        if(y[i] != x[i]):
            ber += 1 

    return(ber/l)
```
